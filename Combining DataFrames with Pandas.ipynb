```python
# Read in first 10 lines of surveys table
survey_sub = surveys_df.head(10)
survey_sub
```

```python
# Import a small subset of the species data designed for this part of the lesson.
# It is stored in the data folder.
species_sub = pd.read_csv('data/speciesSubset.csv', keep_default_na=False, na_values=[""])
species_sub
```

```python
species_sub.columns
```

```python
survey_sub.columns
```

```python
# both contain 'species_id', joining by merge

# inner join
merged_inner = pd.merge(left=species_sub, right=survey_sub,left_on='species_id',right_on='species_id')
# For inner joins, the order of the left and right arguments does not matter.
print(merged_inner.shape)
merged_inner
```

```python
# left join

merged_left = pd.merge(left=survey_sub, right=species_sub,how='left', left_on='species_id',right_on='species_id')
print(merged_left.shape)
merged_left
```

```python
# species_id PF: NaN
merged_left[ pd.isnull(merged_left.genus) ]
```

```python
# right join

merged_right = pd.merge(left=survey_sub, right=species_sub,how='right', left_on='species_id',right_on='species_id')
print(merged_right.shape)
merged_right
```

```python
# Full (outer) join

# the result DataFrame will NaN where data is missing in one of the dataframes. 
# This join type is very rarely used.

merged_out = pd.merge(left=survey_sub, right=species_sub,how='outer', left_on='species_id',right_on='species_id')
print(merged_out.shape)
merged_out
```

```python
# Challenge - Distributions

merged_left=pd.merge(left=surveys_df,right=species_df, how='left', on="species_id")
merged_left=merged_left.groupby('plot_id')['taxa'].nunique()
merged_left.plot(kind='bar')
merged_left
```

```python
# taxa by plot
merged_left=pd.merge(left=surveys_df,right=species_df, how='left', on="species_id")
merged_left=merged_left.groupby(['plot_id','taxa'])['record_id'].count().unstack()
merged_left.plot(kind='bar',stacked=True)
plt.legend(loc='upper center', ncol=3, bbox_to_anchor=(0.5, 1.05))
merged_left
```

```python
# taxa by sex by plot
merged_left=pd.merge(left=surveys_df,right=species_df, how='left', on="species_id")
merged_left['sex']=merged_left['sex'].fillna('X')
merged_left=merged_left.groupby(['plot_id','sex'])['taxa'].count().unstack()
merged_left.plot(kind='bar',stacked=True)
merged_left
```

```python
# Challenge - Diversity Index

plot_info = pd.read_csv("data/plots.csv")
plot_info.groupby("plot_type").count()
```

```python
merged_left
```

```python
merged_plot=pd.merge(left=merged_left,right=plot_info,on='plot_id')
# the number of species for each plot
nspecies = merged_plot.groupby(['plot_id'])['species_id'].nunique()
# the number of individuals
individuals=merged_plot.groupby(['plot_id'])['record_id'].nunique()
# Biodiversity index
biodiversity=pd.concat([nspecies,individuals],axis=1)
biodiversity['diversity_index']=(biodiversity['species_id'])/(biodiversity['record_id'])
```

```python
biodiversity
biodiversity['diversity_index'].plot(kind='bar')
plt.ylabel("diversity index")
plt.xlabel('plot_id')
plt.show()
```
